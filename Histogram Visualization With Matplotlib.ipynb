<h4 style=' font-weight: 600; font-size: 18px; text-align: left'>1.0. Import Required Libraries</h4>

```python
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import numpy as np
np.random.seed(1)
```

<h4 style='font-size: 18px; font-weight: 600'>2.0: Histogram 1 | Basic Histogram</h4>

```python
# Generating Normal Distribution Sample Data
values = np.random.normal(0, 1, 6000)

fig, ax = plt.subplots(figsize=(12,10))
plt.title('Normal Distribution')
plt.hist(values)
fig.savefig(f"./Outputs/Classical Histogram Plot 5.png", bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

<h4 style='font-size: 20px; font-weight: 600'>3.0: Advance Customization</h4>

- Matplotlib has a lot of methods and options to change the appearance of your plot.

-  Styles: Matplotlib has built-in styles and colormaps to choose from. Style is basically how background of your plot would look like. To access all available styles we will use `plt.style.available` method.
-  Colormaps: Colormaps on the other hand are just ranges of colors your plot would use for coloring shapes (bars, dots, lines, etc..)
-  For additional information on colormaps please check <a href="https://matplotlib.org/tutorials/colors/colormaps.html#sphx-glr-tutorials-colors-colormaps-py" target="_blank">official documentation</a>

```python
# Styles available
print( plt.style.available[:20])
```

```python
# Colormaps available
print( plt.colormaps()[:20] ) # Notice we are showing only first 20 available options to keep it clean
```

<h4 style='font-size: 18px; font-weight: 600'>4.0: Histogram 2 | Basic Customization</h4>

```python
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import numpy as np

np.random.seed(1)
values = np.random.normal(0, 1, 6000)

fig, ax = plt.subplots(figsize=(12,10))
fig.patch.set_facecolor("#FFFEFE")
ax.set_facecolor("#FFFFFF")
plt.hist(values, bins=90, facecolor = '#2ab0ff', edgecolor='#169acf', linewidth=0.5)
sns.despine(ax=ax, left=True, right=True, bottom=False)
plt.title('Distribution of Values', fontsize=15)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(linestyle='--', alpha=0.1, color='black')

timestamp = datetime.now().strftime("%Y-%m-%d %H:%M")
data_source = "Simulation"
author = "Jabulente"

text = f"Data Source: {data_source} | Created by: {author} | Generated: {timestamp}"
fig.text(0.5, 0.05, text, ha='center', va='center', fontweight='bold', fontsize=8, color='black')
fig.savefig(f"./Outputs/Classical Histogram Plot 4.png", bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

<h4 style='font-size: 20px; font-weight: 600'>5.0: Histogram 3 | Make It Shine!</h4>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(loc=0, scale=1, size=2000)

fig, ax = plt.subplots(figsize=(12,10))
plt.rcParams.update({'font.family': 'Perpetua'})

n, bins, patches = plt.hist(
    x, bins=90,
    facecolor='#2ab0ff',
    edgecolor='#e0e0e0',
    linewidth=0.5,
    alpha=0.7
)

sns.despine(ax=ax, left=True, right=True, bottom=False)
n = n.astype('int')
for i in range(len(patches)): patches[i].set_facecolor(plt.cm.viridis(n[i] / max(n)))
patches[47].set_fc('red')
patches[47].set_alpha(1)
plt.grid(linestyle='--', alpha=0.1, color='black')

plt.title('Normal Distribution', fontsize=12)
plt.xlabel('Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
fig.savefig(f"./Outputs/Classical Histogram Plot 3.png", bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

<h4 style='font-size: 20px; font-weight: 600'>6.0: Histogram 4 | Add Texts Annotation</h4>

```python
fig, ax = plt.subplots(figsize=(12,10))
plt.rcParams.update({'font.family': 'Perpetua', 'font.style': 'normal', 'font.size': 12})
n, bins, patches = plt.hist(
    values, bins=90, 
    facecolor='#2ab0ff', 
    edgecolor='#e0e0e0', 
    linewidth=0.5, 
    alpha=0.7)

n = n.astype('int') 
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
    

patches[47].set_fc('red')
patches[47].set_alpha(1)

text = 'Comment, Like and Share!'
plt.annotate(
    text, xy=(0.57, 175), 
    xytext=(2, 130), fontsize=10, 
    arrowprops={'width':0.4,'headwidth':7,'color':'#333333'})

sns.despine(ax=ax, left=True, right=True, bottom=False)
plt.title('Variable Distribution', fontsize=12)
plt.xlabel('Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.grid(linestyle='--', alpha=0.1, color='black')
fig.savefig(f"./Outputs/Classical Histogram Plot 2.png", bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```

<h4 style='font-size: 20px; font-weight: 600'>7.0: Histogram 5 | Experiment With Colormap</h4>

```python
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import numpy as np

np.random.seed(1)
values = np.random.normal(0, 1, 6000)

fig, ax = plt.subplots(figsize=(12,10))
fig.patch.set_facecolor("#FEFAFB")
ax.set_facecolor("#FFFDFD")

plt.rcParams.update({'font.family': 'Perpetua', 'font.style': 'normal', 'font.size': 13})
n, bins, patches = plt.hist(values, bins=90, facecolor='#2ab0ff', edgecolor='black', linewidth=0.1, alpha=0.7)
sns.despine(ax=ax, left=True, right=True, bottom=False)
n = n.astype('int')
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.vanimo_r(n[i]/max(n)))
    
patches[47].set_fc('red')
patches[47].set_alpha(1) 
plt.title('Normal Distribution', fontsize=12)
plt.xlabel('Distribution', fontsize=11)
plt.ylabel('Frequency', fontsize=11)
plt.grid(linestyle='--', alpha=0.1, color='black')

text = """Lorem ipsum dolor sit amet, consectetuer 
adipiscing elit. Maecenas porttitor congue massa. ,
magna sed pulvinar ultricies, purus lectus malesuada 
libero, sit amet commodo magna eros quis urna. """

plt.annotate(
    text, xy=(0.57, 175), 
    xytext=(2, 130), fontsize=10, 
    arrowprops={'width':0.4,'headwidth':7,'color':'#333333'})

timestamp = datetime.now().strftime("%Y-%m-%d %H:%M")
data_source = "Simulation"
author = "Jabulente"

text = f"Data Source: {data_source} | Created by: {author} | Generated: {timestamp}"
fig.text(0.5, 0.05, text, ha='center', va='center', fontweight='bold', fontsize=8, color='black')
fig.savefig(f"./Outputs/Classical Histogram Plot.png", bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```
